```python
BASE_DIR = Path.cwd().resolve()
BASE_DIR
```

```python
WEBDRIVER_DIR=BASE_DIR.parent/'centralDriverForSelenium'
```

```python
COOKIE_DIR=BASE_DIR/'cookies'
```

```python
def initialize_selenium():
    global bot
    cookie_directory= str(COOKIE_DIR)
    WEB_DIR = str(WEBDRIVER_DIR)
    options = webdriver.ChromeOptions()
    path_to_chrome_cookie="user-data-dir="+cookie_directory
    options.add_argument(path_to_chrome_cookie)
    bot = webdriver.Chrome(str(WEB_DIR)+'/chromedriver.exe',options=options)
```

```python





initialize_selenium()
```
